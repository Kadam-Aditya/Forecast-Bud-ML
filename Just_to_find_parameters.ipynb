```python
import pandas as pd
import itertools
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
import joblib
```

```python
df = pd.read_csv('budget_dataset.csv')
print(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
ts_data = df['Food']
```

```python
p_values = range(0, 5)  # Replace with your desired range
d_values = range(0, 5)  # Replace with your desired range
q_values = range(0, 5)  # Replace with your desired range

hyperparameter_grid = list(itertools.product(p_values, d_values, q_values))
```

```python
train_size = int(len(ts_data) * 0.8)
train_data, validation_data = ts_data[:train_size], ts_data[train_size:]
```

```python
best_score = float('inf')
best_params = None

for params in hyperparameter_grid:
    p, d, q = params
    model = ARIMA(train_data, order=(p, d, q))
    results = model.fit()

    # Make predictions on the validation set
    forecast = results.get_forecast(steps=len(validation_data))
    predicted_values = forecast.predicted_mean

    # Evaluate model performance (use appropriate metrics for your case)
    mse = mean_squared_error(validation_data, predicted_values)

    # Update best parameters if the current model is better
    if mse < best_score:
        best_score = mse
        best_params = params

print(f"Best hyperparameters: {best_params}")
```

```python
joblib.dump(model, 'parameters.pkl')
```
